## Part 1 ‚Äê Exploratory data analysis

The attached logins.json file contains (simulated) timestamps of user logins in a particular
geographic location. Aggregate these login counts based on 15minute time intervals, and
visualize and describe the resulting time series of login counts in ways that best characterize the

```python
### Preview the Data
logins.head(3)
```

```python
## Check oldest and most recent login_times to see range of time_series 
newest= logins.login_time.max()
oldest = logins.login_time.min()
print('Most Recent: {}'.format(newest))
print("\n")
print('Oldest Date: {}'.format(oldest))
```

**We have 4 months and 12 days of data, the last month is incomplete as it it goes only til the 13th of April**

```python
## Get info to make sure the time series is in datetime format.
logins.info()
```

```python
## We start by setting the index to be login_time then create a login count columns
logins = logins.set_index('login_time')
logins['logins'] = 1
```

```python
logins.head()
```

```python
## We aggregata
logins15 = logins.resample('15Min').count()
logins15.head()
```

```python
## We plot the distribution:
sns.set()
plt.figure(figsize=(20,10))
plt.plot(logins15)
plt.title('Number of Logins between {} & {}'.format(oldest,newest),pad=20, size = 20)
plt.xlabel('Time')
plt.ylabel('Count')
plt.show()
```

We can see some spikes. We need to dig deeper into the data. 

```python
logins15['Hour'] = logins15.index.hour
logins15['Day'] = logins15.index.day
logins15['Weekday'] = logins15.index.day_of_week
logins15['Month'] = logins15.index.month
```

```python
logins15.head()
```

```python
by_hour = logins15.groupby('Hour').logins.sum().reset_index()
by_day = logins15.groupby('Day').logins.sum().reset_index()
by_weekday = logins15.groupby('Weekday').logins.sum().reset_index()
by_month = logins15.groupby('Month').logins.sum().reset_index()
```

```python
## We create visualizations ton show the distribution of logins.

fig, axes = plt.subplots(3,2, figsize =(20,18))
fig.tight_layout(pad=10.0)
fig.suptitle("Distribution of Logins of Particular Region between {} & {}".format(oldest,newest), size = 30)

## By hour of the day.
sns.barplot(ax = axes[0,0], x= 'Hour', y='logins', data = by_hour)
axes[0,0].set_title('Number of logins by hour of the day', size =18, pad =25)

# By day of the Month
sns.lineplot(ax = axes[0,1], x= 'Day', y='logins', data = by_day)
axes[0,1].set_title('Number of logins by day of the month', size =18, pad =25)
axes[0,1].set_ylim(0, 4300)

# By day of the week
sns.barplot(ax = axes[1,0], x= 'Weekday', y='logins', data = by_weekday)
axes[1,0].set_title('Number of logins by day of the week', size =18, pad =25)
axes[1,0].set_xticklabels(['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])

## By week 
axes[1,1].plot(logins15['logins'].resample('W').sum())
axes[1,1].set_title('Number of logins by week', size =18, pad =25)

## By month
sns.barplot(ax = axes[2,0], x= 'Month', y='logins', data = by_month)
axes[2,0].set_title('Number of logins by month', size =18, pad =25)

## By day
axes[2,1].plot(logins15['logins'].resample('D').sum())
axes[2,1].set_title('Number of logins by day', size =18, pad =25)


plt.show()
```
