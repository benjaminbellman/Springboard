**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
## Our boroughs appear to be the index. 
properties_T.index
```

```python
## The column heading for properties_T should be renamed instead of showing integers.
## We can do this by passing the values of the current first row, which are mainly dates (year-month to be precise).
## Then we proceed to delete the first row after we will have used it as our column headers
properties_T.columns = properties_T.iloc[0]
properties_T = properties_T.drop(0)
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
## Let's rename our neighborhoods currently as unnamed: 0 as "location" since at this time, we still have values that are not just neighborhoods.
## Let's also rename the column with strange values as 'location_id'.
properties_T = properties_T.rename(columns={'Unnamed: 0':'location', pd.NaT:'location_id'})
properties_T.head()
```

```python
## We don't really need the location_id column, so let's remove it. 
properties_T = properties_T.drop(columns='location_id')
properties_T.head()
```

Now that we transposed our dataset, we have each date and month as our column headers. We will want to consolidate them and alter our data in the next steps. 

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
## Our dataframe looks better, but we would like for our dates(year-month) and prices to be rows instead of columns. 
## We will use the melt method and keep only our locations in place. 
properties_m = pd.melt(properties_T, id_vars=['location'])
properties_m.head()
```

```python
## We now label the column headings appropriately.
properties_m = properties_m.rename(columns= {0:'date','value':'avg_price'})
properties_m.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
## When checking our properties_m dataframe data types, we notice avg_price is not defined as float64. 
## We will need to find a method to convert it as such. 
properties_m.info()
```

```python
## We  modify the column to convert it to a float: 
properties_m['avg_price'] = pd.to_numeric(properties_m['avg_price'])

## We can check that this method worked. 
properties_m.info()
```

We can currently see that avg_price is missing some values and we may need to remove them since it isn't matching with the amount of non-null values in properties_m[location] and properties_m[date]

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:

## Notice: 
- Since we are in 2021, and we want to analyze our data for the last 20 years, we will look at prices from 2001 - 2021. 

```python
## We create a function whose purpose is to return the price ratio for the borough entered.
def create_price_ratio(x): 
    price_2001 = int(x['avg_price'][x['year']==2001])
    price_2021 = int(x['avg_price'][x['year']==2021])
    ratio = price_2021/price_2001
    return [ratio] 
## Let's call our function on Kensington & Chelsea
create_price_ratio(boroughs_year_price[boroughs_year_price['location']=='Kensington & Chelsea'])
```

```python
## We now need to get all the ratios, so the function will have to be called out multiple times for each borough. 
## Let's start by creating an empty dictionary and calling it ratios. 
## The keys will be the neighborhood, their values will be the ratios. 
ratios = {}
for i in boroughs_year_price['location']:
    borough = boroughs_year_price[boroughs_year_price['location'] == i]
    ratios[i] = create_price_ratio(borough)
ratios
```

```python
## We can make a dataframe out of this dictionary: 
borough_ratios = pd.DataFrame(ratios)
```

```python
## Let's see what it looks like
borough_ratios 
```

```python
## We need to transpose again.
borough_ratios  = borough_ratios .T
borough_ratios .head(3)
```

```python
## We want to reset the index
borough_ratios  = borough_ratios .reset_index()
borough_ratios 
```

```python
#Let's rename the column:
borough_ratios  = borough_ratios .rename(columns={0:'price_ratio','index':'borough'})
borough_ratios .head(5) 
```

```python
## We then sort by Price_Ratio in descending order and return our top 10 
top_10 = borough_ratios.sort_values('price_ratio', ascending = False).head(10)
```

```python
## We can graphically represent the largest ratios of increase for the top 10 boroughs. 
graph_prices = top_10.plot(kind='bar',x='borough',y='price_ratio', title ='Top 15 London Neighborhoods P/R increase from 2001-2021')
graph_prices.set_ylabel('Price Ratio')
graph_prices.set_xlabel('Borough')
plt.show()
```
